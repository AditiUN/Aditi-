```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
books = pd.read_csv('C:/Users/Ashwini/Desktop/ExcelRAssignments/Association Rule/book.csv')
books
```

```python
books.shape
```

```python
df=pd.get_dummies(books)
df.head()
```

# Apriori Algorithm

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
import seaborn as sns
sns.scatterplot( x = rules['support'], y = rules['confidence'], hue= rules['lift'])
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
rules
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules
```

```python
sns.scatterplot(x = rules['support'], y = rules['lift'], hue= rules['confidence'])
```
